```python
import pandas as pd
```

```python
HDW= pd.read_excel("Data\Excels\ATT HDWE CT.xlsx", sheet_name="Design 1")
HDW
```

```python
categories= set(HDW['New Category'])
```

```python
with pd.ExcelWriter("Data\Excels\Database_HDW.xlsx") as writer:
    for category in categories:
        DF= HDW[HDW['New Category']==category]
        DF["Attribute 1"]= None
        DF["Attribute 2"]= None
        DF["Attribute 3"]= None
        DF.to_excel(writer, sheet_name=category, index=False)
```
